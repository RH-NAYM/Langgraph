# Graph 1

---

- The very first langgrap project
- It consists of a `|start point|>node|>End point|`

```python
from tools.VGraph import Helper
__tools__ = Helper()
```

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph

class AgentStateSingleInput(TypedDict): # agent schema
    msg : str

def init_node(state: AgentStateSingleInput) -> AgentStateSingleInput:
    state["msg"] = "Hey "+state.get("msg","")+", Nice learning Langgraph!"
    return state

graph = StateGraph(AgentStateSingleInput)

graph.add_node("init", init_node)

graph.set_entry_point("init")
graph.set_finish_point("init")


app = graph.compile()
__tools__.view_graph(app) # Visualize the graph

result = app.invoke({"msg": "Rakib"})

__tools__.monitor({"data": result})
```

# Multi Input Graph

```python
from typing import TypedDict, List
from langgraph.graph import StateGraph

class AgentStateMultiInput(TypedDict): # Agent Schema
    name: str
    values: list[int]
    result: str

def process_values(state: AgentStateMultiInput) -> AgentStateMultiInput:
    """Process the values and update the result in the state."""

    state["result"] = f"Hi {state['name']}, the sum of your values is {sum(state['values'])}."

    return state


graph = StateGraph(AgentStateMultiInput)

graph.add_node("processor", process_values)

graph.set_entry_point("processor")
graph.set_finish_point("processor")


app = graph.compile()
__tools__.view_graph(app) # Visualize the graph

result = app.invoke(
    {
        "name": "Rakib",
        "values": [1,2,3,4,5,6,7,8,9]
    }
)

__tools__.monitor({"data": result})
```
